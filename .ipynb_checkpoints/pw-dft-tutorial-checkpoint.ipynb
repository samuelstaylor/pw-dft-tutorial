# 1D Plane-Wave Density Functional Theory (DFT) — Toy Model

This notebook implements a **minimal, educational 1D Kohn–Sham DFT solver** using a plane-wave basis. It is designed to help you understand the structure of plane-wave DFT codes, like Quantum ESPRESSO, in a simplified and hackable way.

## Goals
- Understand how **plane waves**, **FFTs**, and **self-consistent field (SCF)** iterations work together.
- Reproduce the **structure** (not full realism) of plane-wave DFT.
- Keep the implementation **minimal** and explicit for learning purposes.

## What We Include
- 1D periodic supercell
- Plane-wave basis with kinetic energy cutoff
- Local external potential (toy pseudopotential)
- Hartree potential (1D Poisson, periodic)
- LDA exchange-correlation (toy functional)
- Self-consistent field (SCF) loop

## What We Omit (on purpose)
- Nonlocal pseudopotentials
- k-point sampling (Gamma only)
- Spin and magnetism
- Advanced mixing (Broyden, DIIS)
- Realistic XC parameterizations

## 1. Import Libraries

````python
import numpy as np
from numpy.fft import fft, ifft, fftfreq
from scipy.linalg import eigh
import matplotlib.pyplot as plt
````

## 2. Parameters, Simulation Cell, Grids, and Initialization

We define the number of electrons, periodic box length, real-space grid, and reciprocal-space grid.

````python
# Control Hartree and XC potentials
TURN_ON_HARTREE = True
TURN_ON_XC = True

# Number of electrons
Ne = 17 # note: max is Ecutoff * 2 + 2

def neutral_ground_state_occupation(Ne):
    n_fully_occupied_band  = Ne // 2
    n_singly_occupied_band = Ne % 2
    occ = [2.0] * n_fully_occupied_band + [1.0] * n_singly_occupied_band
    return occ

occ = neutral_ground_state_occupation(Ne)
print("Occupation numbers:", occ)

# Supercell length (atomic units)
L = 20.0

# Real-space grid
Nx = 512
x = np.linspace(0, L, Nx, endpoint=False)
dx = L / Nx

# Plot dots instead of lines
plt.figure(figsize=(8,4))
plt.scatter(range(len(G)), G, marker='o', color='blue', s=1)  # s = size of dot
plt.xlabel("Index in FFT array")
plt.ylabel("G (reciprocal space wavevector)")
plt.title("Reciprocal-space grid (plane waves)")
plt.tight_layout()
plt.show()
````

## 3. Plane-Wave Cutoff

We define a kinetic energy cutoff $E_{\rm cut}$ in atomic units and select plane waves satisfying $$\frac{|G|^2}{2} \le E_{\rm cut}$$

````python
Ecut = 80.0  # kinetic energy cutoff (a.u.)
pw_mask = 0.5 * G**2 <= Ecut
G_pw = G[pw_mask]
Np = len(G_pw)
print(f"Number of plane waves: {Np}")
````

````python
# -----------------------------
# Plot
# -----------------------------
plt.figure(figsize=(10,4))

# Plot all G points in gray
plt.scatter(range(len(G)), G, marker='o', color='lightgray', s=10, label='All G-points')

# Highlight plane waves within kinetic energy cutoff
plt.scatter(np.where(pw_mask)[0], G_pw, marker='o', color='blue', s=20, label='Selected plane waves (|G|^2/2 ≤ Ecut)')

plt.xlabel("FFT index")
plt.ylabel("G (reciprocal wavevector)")
plt.title("1D Plane-Wave Grid with Kinetic Energy Cutoff")
plt.legend()
plt.tight_layout()
plt.show()
````

## Visualizing Plane Waves in Real Space

In plane-wave DFT, the Kohn–Sham orbitals are expanded in a basis of **plane waves**:

$$
\psi(x) = \sum_G C_G \, e^{i G x}
$$

where \(G\) are the **reciprocal lattice vectors** or **repiprocal-space wavevectors**. They correspond with the plane wave $\phi_{G}(x) = e^{iGx}$. In a 1D periodic supercell of length \(L\) with \(N_x\) grid points, the allowed \(G\) values are:

$$
G = \frac{2 \pi}{L} k, \quad k = -\frac{N_x}{2}, \dots, \frac{N_x}{2}-1
$$

or equivalently in Python:

```python
G = 2 * np.pi * np.fft.fftfreq(Nx, d=L/Nx)
```

Here, $\phi(x) = e^{iGx} = cos(Gx) + i sin(Gx)$ is complex with:

- Real part: $cos(Gx)$

- Imaginary part: $sin(Gx)$
⁡
Plotting these functions in real space helps visualize the oscillatory nature of plane waves. Higher $|G|$correspond to faster oscillations, which are necessary to capture fine features of the density or potential.


````python

# -----------------------------
# Plot a few selected plane waves
# -----------------------------
plt.figure(figsize=(10,6))

# Plot first few positive and negative plane waves
for i, g in enumerate(G_pw[:5]):   # first 10 plane waves
    plt.plot(x, np.cos(g*x), label=f"G={g:.2f}")

plt.xlabel("x (a.u.)")
plt.ylabel("Plane wave amplitude")
plt.title("Real-space view of selected plane waves")
plt.legend()
plt.tight_layout()
plt.show()
````

## 4. External Potential

We use a smooth Gaussian potential as a toy local pseudopotential:
$$V_{\rm ext}(x) = V_0 \exp\Big[-\frac{(x-x_0)^2}{2\sigma^2}\Big]$$

````python
V0 = -5.0
sigma = 0.5
x0 = L / 2

V_ext = V0 * np.exp(-((x - x0)**2)/(2*sigma**2))
V_ext_G = fft(V_ext) / Nx
````

## 5. Hartree Potential (1D)

The Hartree potential satisfies the 1D Poisson equation for periodic density:
$$\frac{d^2 V_H}{dx^2} = -4\pi n(x)$$
and in reciprocal space:
$$V_H(G) = \frac{4\pi n(G)}{G^2}, \quad G \ne 0$$

````python
def hartree_potential(n):
    nG = fft(n)
    VH_G = np.zeros_like(nG, dtype=complex)
    if not TURN_ON_HARTREE:
        return np.zeros_like(n)
    for i, g in enumerate(G):
        if abs(g) > 1e-12:
            VH_G[i] = 4.0 * np.pi * nG[i] / (g**2)
    VH = np.real(ifft(VH_G))
    return VH
````

## 6. Exchange-Correlation Potential (Toy LDA)

We define a simple LDA-like functional:
$$E_{xc}[n] = -A \int n(x)^{4/3} dx, \quad V_{xc} = \frac{dE_{xc}}{dn} = -\frac{4}{3} A n^{1/3}$$

````python
A_xc = 0.75
def v_xc(n):
    if TURN_ON_XC:
        return -(4/3) * A_xc * np.maximum(n, 1e-10)**(1/3)
    return np.zeros_like(n)
````

## 7. Kohn–Sham Hamiltonian

The Hamiltonian in plane-wave basis is:
$$H_{GG'} = \frac{G^2}{2} \delta_{GG'} + V_{\rm eff}(G-G')$$

````python
def ks_hamiltonian(Veff):
    Veff_G = fft(Veff) / Nx
    H = np.zeros((Np, Np), dtype=complex)
    for i, Gi in enumerate(G_pw):
        H[i, i] = 0.5 * Gi**2
        for j, Gj in enumerate(G_pw):
            dG = Gi - Gj
            idx = np.where(np.isclose(G, dG))[0]
            if idx.size > 0:
                H[i, j] += Veff_G[idx[0]]
    return H
````

## 8. Density from Orbitals

Compute the electron density from Kohn–Sham orbitals $\psi_i(x)$ and occupations $f_i$:
$$n(x) = \sum_i f_i |\psi_i(x)|^2$$

````python
def density_from_orbitals(C, occ):
    n = np.zeros(Nx)
    for i in range(len(occ)):
        psiG = np.zeros(Nx, dtype=complex)
        psiG[pw_mask] = C[:, i]
        psi = ifft(psiG)  # * Nx optional depending on FFT normalization
        n += occ[i] * np.abs(psi)**2
    return n
````

## 9. Self-Consistent Field (SCF) Loop

Iterate until density converges:

````python
# Initial guess: uniform density
n = np.ones(Nx) * (Ne / L)

alpha = 0.3  # mixing parameter
n_iter = 50
conv_thr = 1e-6

# Track convergence
delta_list = []

for it in range(n_iter):
    # Compute potentials
    VH = hartree_potential(n)
    Vxc = v_xc(n)
    Veff = V_ext + VH + Vxc
    
    # Build Hamiltonian and diagonalize
    H = ks_hamiltonian(Veff)
    eigvals, eigvecs = eigh(H)
    
    # Compute new density
    n_new = density_from_orbitals(eigvecs, occ)
    
    # Compute density difference
    delta = np.linalg.norm(n_new - n) * dx
    delta_list.append(delta)
    
    # Mix densities
    n = alpha * n_new + (1 - alpha) * n
    
    print(f"Iteration {it:3d}: E0 = {eigvals[0]:.6f}   delta_n = {delta:.3e}")
    
    if delta < conv_thr:
        print(f"SCF converged in {it+1} iterations (delta_n < {conv_thr})")
        break

# -----------------------------
# Convergence Plot
# -----------------------------
plt.figure(figsize=(6,4))
plt.semilogy(range(1, len(delta_list)+1), delta_list, marker='o')
plt.xlabel("SCF Iteration")
plt.ylabel(r"$\Delta n$ (L2 norm)")
plt.title("SCF Convergence of Electron Density")
plt.grid(True, which="both", ls="--")
plt.tight_layout()
plt.show()
````

## 10. Plot Final Electron Density

````python
plt.figure(figsize=(8,4))
plt.plot(x, n, label="Electron density")
plt.plot(x, V_ext / np.max(np.abs(V_ext)) * np.max(n), '--', label="V_ext (scaled)")
plt.xlabel("x")
plt.ylabel("Density")
plt.title("1D Plane-Wave DFT: Self-Consistent Density")
plt.legend()
plt.tight_layout()
plt.show()
````

## 11. Plot Kohn–Sham Potential Components

We can visualize $V_{\rm ext}$, $V_H$, $V_{xc}$, and $V_{\rm eff}$.

````python
SCALE = True

plt.figure(figsize=(8,5))
plt.plot(x, V_ext, label="V_ext")
if SCALE:
    plt.plot(x, VH / np.max(np.abs(VH)), label="V_Hartree (scaled)")
    plt.plot(x, Vxc / np.max(np.abs(Vxc)), label="V_xc (scaled)")
else:
    plt.plot(x, VH, label="V_Hartree")
    plt.plot(x, Vxc, label="V_xc")

plt.plot(x, Veff, '--', linewidth=2, label="V_eff = V_ext + V_H + V_xc")
plt.xlabel("x")
plt.ylabel("Potential (a.u.)")
plt.title("Kohn–Sham Potential Components")
plt.legend()
plt.tight_layout()
plt.show()
````
